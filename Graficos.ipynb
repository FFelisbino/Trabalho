```python
import numpy as np

dados = np.random.randn(1000)

fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].hist(dados, bins=10, alpha =0.5, color='blue', edgecolor='black')
axs[0].set_title('10 Bins')

axs[1].hist(dados, bins=30, alpha =0.5, color='blue', edgecolor='black')
axs[1].set_title('30 Bins')

axs[2].hist(dados, bins=50, alpha =0.5, color='blue', edgecolor='black')
axs[2].set_title('50 Bins')

for ax in axs:
  ax.set_xlabel('Valores')
  ax.set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
plt.hist(df['nota'], edgecolor='black', linewidth=1.9, bins=5) #bins = nº de intervalos
plt.show()
```

```python
#definir a largura da barra
bin_windth = 5

#definir onde começa a primeira barra e onde termina a última barra
start_bin = 0
end_bin = 10
min_val = min(df['nota'])
max_val = max(df['nota'])

bins = []
atual = start_bin
while atual <= end_bin:
  bins.append(atual)
  atual += bin_windth
plt.hist(df['nota'], edgecolor='black', linewidth=1.2, bins = bins)
plt.show()
```

```python
valores = df['cargo'].value_counts()
labels = valores.index
#unique x index -> unique = verifica diversidade de valores em uma coluna. Index = estrutura que representa linhas e colunas
plt.pie(valores, labels = labels)
plt.title('Gráfico de pizza')
plt.show()
```

```python
valores = df['cargo'].value_counts()
labels = df['cargo'].value_counts().index

plt.pie(valores, labels=labels, autopct='%1.1f%%') #autopct = porcentagem
plt.title('Gráfico com porcentagem')
plt.show()
```

```python
explode = [0, 0, 0.1, 0]
valores = df['cargo'].value_counts()
labels = valores.index

plt.pie(valores, labels=labels, autopct='%1.1f%%', explode=explode) #autopct = porcentagem
plt.title('Gráfico com porcentagem')
plt.show()
```

```python
#Plotar as diferenças entre sálarios por estados
salario_por_estado = df.groupby('estado')['salario'].mean().reset_index()

plt.figure(figsize=(5,6)) # ajustar o tamanho da figura
plt.bar(salario_por_estado['estado'], salario_por_estado['salario'])
plt.xlabel('Estado')
plt.ylabel('Salário Médio')
plt.title('Salário Médio por Estado')
plt.xticks(rotation=45) # rotacionar os rótulos do
```

```python
#salário de produção de SP e MG
df_producao = df[(df['departamento'] == 'Produção') & (df['estado'].isin(['SP', 'MG']))]

salario_por_estado = df_producao.groupby('estado')['salario'].mean().reset_index()

plt.figure(figsize=(8,6)) # ajustar o tamanho da figura
plt.bar(salario_por_estado['estado'], salario_por_estado['salario'])
plt.xlabel('Estado')
plt.ylabel('Salário Médio')
plt.title('Salário Médio por Estado')
```
