```python
#definir a largura da barra
bin_windth = 5

#definir onde começa a primeira barra e onde termina a última barra
start_bin = 0
end_bin = 10
min_val = min(df['nota'])
max_val = max(df['nota'])

bins = []
atual = start_bin
while atual <= end_bin:
  bins.append(atual)
  atual += bin_windth
plt.hist(df['nota'], edgecolor='black', linewidth=1.2, bins = bins)
plt.show()
```
